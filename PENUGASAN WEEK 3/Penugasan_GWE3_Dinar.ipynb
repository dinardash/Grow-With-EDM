## BAB V: Multivariate Analysis & Correlation

### Soal 11: Bivariate - Kategori vs Numerik
**Deskripsi:** Bisnis ingin tahu kategori produk mana yang paling menguntungkan. Buat visualisasi Bar Chart yang menampilkan total Profit untuk setiap `Category`.
**Hint:** Lakukan `.groupby('Category')['Profit'].sum().reset_index()` terlebih dahulu, lalu hasilnya di-plot dengan barplot.

```python
# Tulis kode untuk Soal 11 di sini
df.groupby('Category')['Profit'].sum().reset_index().plot(
    kind='bar',
    x='Category',
    y='Profit'
)
```

**Tuliskan Analisis Output Soal 11:**
- *Berdasarkan bar chart di atas, dapat disimpulkan bahwa semakin banyak penjualan pada suatu kategori, maka semakin banyak profit yang akan di dapatkan. Hal ini dapat dilihat dari urutan jumlah penjualan dan profit yang sama

### Soal 12: Bivariate - Numerik vs Numerik (Scatter Plot)
**Deskripsi:** Buat Scatter Plot untuk melihat hubungan antara `Sales` dan `Profit`.
**Hint:** Gunakan `sns.scatterplot()`.

```python
# Tulis kode untuk Soal 12 di sini
sns.scatterplot(data=df, x="Sales", y="Profit")
```

**Tuliskan Analisis Output Soal 12:**
- Scatter plot menunjukkan hubungan positif yang cenderung linear antara Sales dan Profit. Namun, penyebaran data semakin lebar seiring meningkatnya Sales. ini berarti bahwa walaupun jumlah penjualan sama, jumlah profit yang dihasilkan dapat berbeda. perbedaan jumlah profit ini dapat disebabkan oleh faktor lain, seperti kategori dan margin keuntungan untuk setiap produk.

### Soal 13: Pair Plot (Pemetaan Masif)
